```python
class clssimp(torch.nn.Module):
    def __init__(self, ch=2048, num_classes=81):

        super(clssimp, self).__init__()
        self.pool = torch.nn.AdaptiveAvgPool2d(output_size=(1, 1))
        self.way1 = nn.Sequential(
            nn.Linear(ch, 1000, bias=True),
            nn.BatchNorm1d(1000),
            nn.ReLU(inplace=True),
        )
        self.cls= nn.Linear(1000, num_classes,bias=True)

    def forward(self, x):
        # bp()
        x = self.pool(x)
#         print(f'x.shape={x.shape}')
        x = x.reshape(x.size(0), -1)
        x = self.way1(x)
        logits = self.cls(x)
        return logits
    
    def intermediate_forward(self, x):
        x = self.pool(x)
        x = x.reshape(x.size(0), -1)
        x = self.way1(x)
        return x
```

```python
def clsi_forward(self, x):
        # bp()
        #print(self)
        x = self.pool(x)
        #print(x)
        x = x.reshape(x.size(0), -1)
        x = self.way1(x)
        logits = self.cls(x)
        return logits

def clsi_intermediate_forward(self, x):
    x = self.pool(x)
    x = x.reshape(x.size(0), -1)
    x = self.way1(x)
    return x
```

```python
img_shape=(3, ground_truth_1.shape[2], ground_truth_1.shape[3])
input_gradient1=[]
gr = ground_truth_1.clone()
a_i = 0

for i in range(len(label_true)):
    a_i = i
    output, stats = rec_machine.reconstruct(input_gradient, label_true[i], input_gradient1, a_i,  ca, canny_reg2, gr, img_shape)
    #print(output)
    final_out.append(output.clone())
    output1 = output.clone()
#     output1 = output.mul(ds).add(dm).clamp_(0, 1)
    max_out = torch.max(torch.tensor(output1))
    input_gradient1= output1.expand(config['restarts'], 1, img_shape[0],img_shape[1],img_shape[2])
    
    im = output1.mul(255).byte()
    im = im.cpu().numpy().squeeze(0).transpose((1, 2, 0))
    
    output2=cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im1 = ground_truth_1.mul(255).byte()
    im2 = im1.cpu().numpy().squeeze(0).transpose((1, 2, 0))
    ground_truth_2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
    
    ssim, s = structural_similarity(output2, ground_truth_2, full=True)
    
    test_mse = (output1.detach() - ground_truth_1).pow(2).mean()
    feat_mse = (model(output1.detach())- model(ground_truth_1)).pow(2).mean()  
    test_psnr = inversefed.metrics.psnr(output1, ground_truth_1, False, max_out)
   
    
    lpips = lf_LPIPS.forward(output1,ground_truth_1)
    
    final_state[i] = stats['opt']
                    #final_state.append(stats['opt'])
    final_mse[i] = test_mse
    final_feat_mse[i] = feat_mse
    final_psnr[i] = test_psnr
    final_ssim[i] = ssim
    final_lpips[i] = lpips
```
